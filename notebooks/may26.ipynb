### Summary

Plots below are constructed by:

1) finding the timestep closest to 300 Nperiods (about 5 hours) for each run

2) getting wprime and thetaprime values for all 10 ensembles for that timestep (i.e. 10 x 312 x 128 x 192)

3) getting 4 wprime xthetaprime vertical profiles by conditionally sampling at each level over the 10 members and 128x192 horizontal grid points

5) the legend for these is of the form: wp_tp = positive w, positve theta, wp_tn = positive w, negative theta, etc.  plus their sum (the net heat flux)

6) the two horizontal lines show zf0 (base of the entrainment layer for zero net heat flux (note that if this doesn't quite correspond to zero on the x axis it's because of layer discretization)  and zg0 (bottom line, Niamh's gradient threshold)


###  What I was expecting to see:  

I associated entrainment with downward moving warm air, so I thought the biggest differences would be in the profiles of wn_tp  

### What the graphs show

In contrast, all of the action (for both zf0 and zg0) is occuring in the upward moving buoyant parcels (wp_tp)
